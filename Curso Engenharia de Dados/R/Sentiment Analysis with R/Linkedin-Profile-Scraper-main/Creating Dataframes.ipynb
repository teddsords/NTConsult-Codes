```python
import pandas as pd
```

```python

df_contratados = pd.read_csv('contratados.csv')
df_contratados.head(3)
```

```python
df_contratados['contratados'] = 1
df_contratados.head(3)
```

```python
df_nao_contratados = pd.read_csv('nao contratados.csv')
df_nao_contratados.head(3)
```

```python
df_nao_contratados['contratados'] = 0
df_nao_contratados.head(3)
```

```python
print(df_nao_contratados.info())
print(df_contratados.info())
```

```python
perfis_linkedin = pd.concat([df_contratados, df_nao_contratados])
perfis_linkedin.head(3)
```

```python
perfis_linkedin.to_csv('amostra_de_perfis_de_candidatos_NTConsult_2022.csv')
```
