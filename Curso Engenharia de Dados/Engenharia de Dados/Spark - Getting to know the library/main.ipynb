# Converting String -> Double

In this section we will be converting columns with String data type to Double when necessary

```python
empresas.limit(5).toPandas()
```

```python
from pyspark.sql.types import DoubleType, StringType
from pyspark.sql import functions as f
```

```python
empresas.printSchema()
```

```python
empresas = empresas.withColumn('capital_social_da_empresa',
             f.regexp_replace('capital_social_da_empresa', ',', '.'))

empresas.limit(5).toPandas()
```

```python
empresas = empresas.withColumn('capital_social_da_empresa', empresas['capital_social_da_empresa'].cast(DoubleType()))
empresas.limit(5).toPandas()
```

```python
empresas.printSchema()
```

# Converting String->Date

In this section we will be converting Strings into Dates

```python
estabelecimentos.limit(5).toPandas()
```

```python
estabelecimentos = estabelecimentos\
    .withColumn(
        'data_situacao_cadastral',
        f.to_date(estabelecimentos.data_situacao_cadastral.cast(StringType()), 'yyyyMMdd')
    )\
    .withColumn(
        'data_de_inicio_atividade',
        f.to_date(estabelecimentos.data_de_inicio_atividade.cast(StringType()), 'yyyyMMdd')
    )\
    .withColumn(
        'data_da_situcao_especial',
        f.to_date(estabelecimentos.data_da_situacao_especial.cast(StringType()), 'yyyyMMdd')
    )

estabelecimentos.printSchema()
```

```python
socios.limit(5).toPandas()
```

```python
socios = socios.withColumn('data_de_entrada_sociedade', 
                            f.to_date(socios.data_de_entrada_sociedade.cast(StringType()), 'yyyyMMdd'))

socios.printSchema()
```

# Identifying null values

In this section we will learn how to identify Null values within a DataFrame

```python
socios.limit(5).toPandas()
```

```python
socios.limit(5).show()
```

```python
'''
 Para fazer a contagem de quantos nulls temos nas colunas podemos utilizar o seguinte código
 Criamos o nosso comando Select, onde utilizamos a list comprehension para ter os nomes das colunas
 do DataFrame. Então para cada coluna no DataFrame, vamos contar todas as ocorrências quando 
 a coluna contém um valor de null
'''

socios.select(
    [f.count(f.when(f.isnull(c), 1)).alias(c) for c in socios.columns]
).show()
```

```python
# Para preencher esses dados devemos notar que as colunas são de diferentes tipos, por exemplo:
# pais é do tipo integer e nome do representante é string

socios.na.fill(0).limit(5).toPandas()
```

```python
socios.na.fill('-').limit(5).toPandas()
```

# Ordering Data

In this section we will learn how to order data using SQL commands.

```python
socios.select(
    'nome_do_socio_ou_razao_social', 'faixa_etaria',
    f.year('data_de_entrada_sociedade').alias('ano_de_entrada')
).orderBy('ano_de_entrada', ascending=False).show(5)
```

```python
socios.select(
    'nome_do_socio_ou_razao_social', 'faixa_etaria',
    f.year('data_de_entrada_sociedade').alias('ano_de_entrada')
).orderBy(['ano_de_entrada', 'faixa_etaria'], ascending=[False, False]).show(10)
```

# Filtering Data

In this section we will learn how to filter data

```python
empresas.where(
    'capital_social_da_empresa= 50'
).show(5, False)
```

```python
socios\
    .select('nome_do_socio_ou_razao_social')\
    .filter(socios.nome_do_socio_ou_razao_social.startswith('RODRIGO'))\
    .filter(socios.nome_do_socio_ou_razao_social.endswith('DIAS'))\
    .limit(10).toPandas()
```

# Like command

In this section we will learn more about the like command

```python
empresas.select(
    'razao_social_nome_empresarial', 'natureza_juridica', 'porte_da_empresa', 'capital_social_da_empresa'
).filter(
    f.upper(empresas['razao_social_nome_empresarial']).like('%RESTAURANTE')
).show(15, False)
```
